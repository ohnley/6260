```python
# Visualize some similarities between words
def gen_sim_matrix(words):
    n = len(words)
    similarity_matrix = np.zeros((n, n))
    
    for i in range(n):
        for j in range(n):
            if not ((words[i] in vectors) and (words[j] in vectors)): similarity_matrix[i, j] = None
            elif (i==j): similarity_matrix[i, j] = None
            else: similarity_matrix[i, j] = vectors.similarity(words[i], words[j])
    
    df = pd.DataFrame(similarity_matrix, index=words, columns=words)
    
    # Plot the similarity matrix as a heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(df, annot=True, cmap='cool', vmin=0, vmax=1)
    plt.title('Word Similarity Matrix')
    plt.tight_layout()
    plt.show()
```

```python
words = ['hippo','swampy', 'river', 'mammal', 'cat', 'bread', 'water', 'donut', 'freezer', 'brain', 'oily']
gen_sim_matrix(words)
```

```python
words_2 = ['cactus','pyramid', 'camel', 'pharaoh','octagon', 'machine', 'desert', 'dessert', 'ice_cream']
gen_sim_matrix(words_2)
```
